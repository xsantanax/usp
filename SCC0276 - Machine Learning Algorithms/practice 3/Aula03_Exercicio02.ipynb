*   Com base na variáveis escolhida via matriz de correlação, aproxime a função por uma reta e calcule seu erro quadrático médio

* Dica 1: Utilize os pacotes importados para realizar a aproximação por função linear e para calcular o erro quadrático médio. 
* Dica 2: Para criar um novo conjunto com a variável selecionada,  inicialize um `DataFrame()` do pandas passando como parâmetro as colunas do DataFrame original concatenadas com a função `_c` da biblioteca numpy:

````
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


# x = variável mais relevante -> DF[0] (ou 12?) Verificar texto da questão anterior
# y = saída -> DF[14]

# Por que concatenar?????
# Vou fazer sem concatenar

x = DF[0]
y = DF[14]

x2d = x[:,np.newaxis] #x precisa ser 2D na função LinearRegression()

lr = LinearRegression().fit(x2d,y) 
#lr.score(x2d,y)

y_predito = lr.predict(x2d)

fig = plt.figure()
plt.plot(x, y, 'g.')
plt.plot(x, y_predito, 'b-')
plt.title('Regressão Linear - 1 Variável')
plt.show()

# Erro quadrático médio
EQM = mean_squared_error(y, y_predito)
print("Erro Quadrático Médio:")
print(EQM)
````

````
#Agora, tentando com 2 variaveis

x = pd.DataFrame(np.c_[DF[0], DF[12]], columns = ['Variavel 0', 'Variavel 12'])
y = DF[14]


lr = LinearRegression().fit(x,y) 

#print(lr.score(x,y))

y_predito = lr.predict(x)

fig = plt.figure()
plt.plot(x, y, 'g.')
plt.plot(x, y_predito, 'b-')
plt.title('Regressão Linear - 2 Variáveis')
plt.show()

# Erro quadrático médio
EQM = mean_squared_error(y, y_predito)
print("Erro Quadrático Médio:")
print(EQM)

````

*  Agora calcule uma aproximação por uma função quadrática. Verifique qual produz o menor erro quadrático médio: a aproximação via função linear ou via função quadrática

  * Dica: Utilize a biblioteca já importada para *transformar* os dados para uma regressão polinomial (func. quadrática). Mais informações sobre como o processo podem ser encontradas dentro da pasta do projeto.



````
from sklearn.preprocessing import PolynomialFeatures

x = pd.DataFrame(np.c_[DF[0], DF[12]], columns = ['Variavel 0', 'Variavel 12'])
y = DF[14]

np.set_printoptions(formatter={'float': lambda x: "{0:0.2f}".format(x)}) #imprime apenas 2 casas decimais


# Utilizando um polinomio de grau 2
p = PolynomialFeatures(degree=2)

X_Transf = p.fit_transform(x)
Nomes = p.get_feature_names()
#print(Nomes)
#print(X_Transf)

x_ajustado = pd.DataFrame(np.c_[X_Transf], columns=Nomes)
#print(x_ajustado)

lr = LinearRegression().fit(x_ajustado,y) 
y_predito = lr.predict(x_ajustado)
#print(y_predito)


fig = plt.figure()
plt.plot(x, y, 'g.')
plt.plot(x, y_predito, 'b-')
plt.title('Regressão Linear - 2 Variáveis - Polinomial de grau 2')
plt.show()


#lr = LinearRegression().fit(X_Transf,y) 
#y_predito = lr.predict(x)

# Erro quadrático médio
EQM = mean_squared_error(y, y_predito)
print("Erro Quadrático Médio:")
print(EQM)
````

````
#Agora, fazendo a regressão polinomial de grau 2 com apenas uma variável

# x = variável mais relevante -> DF[0] (ou 12?) Verificar texto da questão anterior
# y = saída -> DF[14]

# Por que concatenar?????
# Vou fazer sem concatenar

x = DF[0]
y = DF[14]

x2d = x[:,np.newaxis] #x precisa ser 2D na função LinearRegression() e fit_transform()


# Utilizando um polinomio de grau 2
p = PolynomialFeatures(degree=2)

X_Transf = p.fit_transform(x2d)
Nomes = p.get_feature_names()
print(Nomes)
print(X_Transf)

x2d_ajustado = pd.DataFrame(np.c_[X_Transf], columns=Nomes)
#print(x2d_ajustado)

lr = LinearRegression().fit(x2d_ajustado,y) 
y_predito = lr.predict(x2d_ajustado)
#print(y_predito)


fig = plt.figure()
plt.plot(x, y, 'g.')
plt.plot(x, y_predito, 'b-')
plt.title('Regressão Linear - 1 Variável - Polinomial de grau 2')
plt.show()

# Erro quadrático médio
EQM = mean_squared_error(y, y_predito)
print("Erro Quadrático Médio:")
print(EQM)
````

*   Por fim, calcule uma aproximação por uma função cúbica. Verifique qual produz o menor erro quadrático médio: a aproximação via função linear, via função quadrática ou via função cúbica.


````
x = DF[0]
y = DF[14]

x2d = x[:,np.newaxis] #x precisa ser 2D na função LinearRegression() e fit_transform()


# Utilizando um polinomio de grau 3
p = PolynomialFeatures(degree=3)

X_Transf = p.fit_transform(x2d)
Nomes = p.get_feature_names()
print(Nomes)
print(X_Transf)

x2d_ajustado = pd.DataFrame(np.c_[X_Transf], columns=Nomes)
#print(x2d_ajustado)

lr = LinearRegression().fit(x2d_ajustado,y) 
y_predito = lr.predict(x2d_ajustado)
#print(y_predito)


fig = plt.figure()
plt.plot(x, y, 'g.')
plt.plot(x, y_predito, 'b-')
plt.title('Regressão Linear - 1 Variável - Polinomial de grau 3')
plt.show()

# Erro quadrático médio
EQM = mean_squared_error(y, y_predito)
print("Erro Quadrático Médio:")
print(EQM)
````

A aproximação via função cúbica produziu o menor erro quadrático médio.
