# PCA
 O conjunto Iris consiste na medida de 4 características de plantas e seus respectivos tipos. Embora não seja possível visualizar o conjunto de 4 dimensões, podemos utilizar a técnica PCA para reduzir a dimensionalidade do conjunto para 2 ou 3 dimensões.
- Utilize a técnica PCA para visualizar em 2 dimensões o conjunto Iris, já carregado abaixo. 
Lembre-se que os dados devem estar normalizados e centrados na média antes de aplicar o PCA.
  - Dica: Utilize a classe StandardScaler para pré-processar os dados

```
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn import datasets

iris = datasets.load_iris()
x = iris.data
y = iris.target


#Pré-processamento (Média = 0, variância = 1)
scaler = StandardScaler()
scaler.fit(x)
#print(scaler.mean_)
x_ajustado = scaler.transform(x)
#print(x_ajustado)


#####
#plt.figure()
#plt.scatter(x_ajustado[:, 0], x_ajustado[:, 1], c=y, cmap=plt.cm.Set1,edgecolor='k') #Por que as diferentes cores??
#####


#matriz de covariancia
#print(np.cov(x_ajustado))
#print(np.corrcoef(x_ajustado)) #tem diagonal 1
mcov = np.cov(x_ajustado)


#autovalores e autovetores
autovalores,autovetores = np.linalg.eig(mcov)
#print(autovalores[0])
#print(len(autovalores))
#print(autovetores[0])
#print(len(autovetores[0]))
#150 autovalores e 150 autovetores de 150 dimensões

#Atribuindo os autovetores de maior autovalor para as componentes principais Z1 e Z2
DF = pd.DataFrame({'autovalores':autovalores, 'index':np.arange(150)})
DFordenado = DF.sort_values(by='autovalores', axis='index', ascending=False)
Z1 = (autovetores[DFordenado.index[0]])
Z2 = (autovetores[DFordenado.index[1]])
#print(DFordenado)
#print (Z1[0])
#print (Z2[0])
#print (autovetores)
#np.cov(Z1,Z2)


#####
#plt.figure()
#plt.scatter(Z1, Z2, c=y, cmap=plt.cm.Set1,edgecolor='k') #Por que as diferentes cores??
#####
```



---

- Mostre a variância e a porcentagem da variância que é expressa por cada componente principal.
  - Dica: Utilize as variáveis `explained_variance_`  e `explained_variance_ratio_`  da classe PCA. 

```
pca = PCA(n_components=2)
pca.fit(Z1,Z2)
print(pca.explained_variance)  
print(pca.explained_variance_ratio_)  
```
