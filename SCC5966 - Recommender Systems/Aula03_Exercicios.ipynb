***Exercício 02:*** Como visto, o algoritmo ItemAttributeKNN pode ser usado com diferentes tipos de metadados, como gêneros, tags e palavras no geral. Mais do que isso, podemos adaptá-lo também para que a similaridade entre itens seja feita com base em informações multimídia, como imagens, áudio, etc.

A base de dados utilizada até o momento, ml-20m-compact.tar.gz possui, além das interações de usuários com filmes, uma série de arquivos que contém informações multimídia que foram extraídas dos trailers de cada filme. Esses arquivos estão condensados no zip ml-20m-features.tar.gz, o qual foi feito o download e extraído acima.

Considere por exemplo o arquivo visual_histogram_50.arq. Ele possui 433 vetores (no. de filmes) de tamanho 50. Podemos pensar que cada vetor desse representa informações visuais (cor, brilho, imagem, etc.) que foram extraídas dos trailers de cada filme.

Sua tarefa é usar esses vetores de características visuais no cálculo de similaridade entre os filmes, e em seguida, aplicar essas similaridades no algoritmo ItemAttributeKNN para gerar recomendações.

Dica 1: para calcular a similaridade entre dois vetores pode-se usar o ângulo de cosseno (vide https://en.wikipedia.org/wiki/Cosine_similarity).

Dica 2: é possível passar para o algoritmo ItemAttributeKNN a matriz de similaridade entre itens, por meio do parâmetro similarity_file=arquivo. Veja em: https://github.com/caserec/CaseRecommender/blob/master/caserec/recommenders/rating_prediction/item_attribute_knn.py

```python
# TODO
from sklearn.metrics.pairwise import cosine_similarity
from caserec.recommenders.rating_prediction.item_attribute_knn import ItemAttributeKNN

# Load the visual features
with open('./features/visual_histogram_50.arq', 'rb') as arq_visualHistograms:
    visualHistograms = pickle.load(arq_visualHistograms)

# Calculate cosine similarity between visual features vectors
visual_similarity_matrix = cosine_similarity(visualHistograms)

# Round the similarity matrix values to integers
visual_similarity_matrix_rounded = np.round(visual_similarity_matrix).astype(int)

# Save the rounded similarity matrix to a file with tab as the separator
np.savetxt('visual_similarity_matrix.txt', visual_similarity_matrix_rounded, delimiter='\t', fmt='%d')

# Use ItemAttributeKNN with visual similarity as metadata
recommender = ItemAttributeKNN('train.dat', 'test.dat', similarity_file='visual_similarity_matrix.txt', as_similar_first=True, k_neighbors=5)

# Compute recommendations
recommender.compute()
```

***Exercício 03:*** Implementar uma função que retorna a probabilidade de um item ser relevante para um usuário, considerando os gêneros dos filmes. Utilize métodos probabilísticos.
- A partir do conjunto de treinamento, obter todas as interações do usuário u.
- Rotular as notas desse usuário como: item relevante se nota >=3 e não relevante se nota < 3.
- Dado um item do conjunto de teste, aplicar o Teorema de Bayes com suavização de Laplace. Utilizar os gêneros associados.
- Retornar se o item é ou não relevante, e em seguida, comparar o resultado com a nota real que esse usuário deu para o item (disponível no conjunto de teste).

```python
# TODO
import pandas as pd
import numpy as np

# Build movies_genres
movies_genres = movies.drop('genres', axis=1).join(movies.genres.str.split('|', expand=True)
             .stack().reset_index(drop=True, level=1).rename('genre'))
movies_genres.dropna(inplace=True)
movies_genres['movieId'] = movies_genres.movieId.astype(int)

# Extract unique genres from the 'movies_genres' DataFrame
genres_unique = movies_genres['genre'].unique()

# Step 1: Calculate relevance probabilities for each genre based on training ratings
def calculate_genre_probabilities(train):
    genre_probabilities = {}
    total_ratings = len(train)

    for genre in genres_unique:
        genre_ratings = train[train['movieId'].isin(movies_genres[movies_genres['genre'] == genre]['movieId'])]
        relevant_ratings = genre_ratings[genre_ratings['rating'] >= 3]
        probability = (len(relevant_ratings) + 1) / (len(genre_ratings) + 2)  # Laplace smoothing
        genre_probabilities[genre] = probability

    return genre_probabilities

# Step 2: Calculate the probability of an item being relevant to a user based on the item's genres
def calculate_item_probability(user_id, item_genres, genre_probabilities):
    user_ratings = train[train['userId'] == user_id]
    user_relevant_genres = set()

    for index, row in user_ratings.iterrows():
        movie_genres = movies_genres[movies_genres['movieId'] == row['movieId']]['genre'].tolist()
        user_relevant_genres.update(movie_genres)

    item_probability = 1.0

    for genre in item_genres:
        if genre in user_relevant_genres:
            item_probability *= genre_probabilities[genre]
        else:
            item_probability *= (1 - genre_probabilities[genre])

    return item_probability

# Step 3: Compare the calculated probability with the user's actual rating in the test set
def evaluate_relevance_predictions(test, genre_probabilities):
    correct_predictions = 0
    total_predictions = len(test)

    for index, row in test.iterrows():
        user_id = row['userId']
        item_genres = movies_genres[movies_genres['movieId'] == row['movieId']]['genre'].tolist()
        item_probability = calculate_item_probability(user_id, item_genres, genre_probabilities)

        if item_probability >= 0.5 and row['rating'] >= 3:
            correct_predictions += 1
        elif item_probability < 0.5 and row['rating'] < 3:
            correct_predictions += 1

    accuracy = correct_predictions / total_predictions
    return accuracy


# Calculate the relevance probabilities for each genre based on training data
genre_probabilities = calculate_genre_probabilities(train)

# Evaluate the accuracy of relevance predictions on the test set
accuracy = evaluate_relevance_predictions(test, genre_probabilities)
print(f'The accuracy of relevance predictions is {accuracy * 100:.2f}%')
```
